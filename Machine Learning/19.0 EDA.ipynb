# EDA extended

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df.drop(['Id'], axis=1, inplace=True)
df.head()
```

```python
sns.distplot(df.SepalLengthCm)
```

```python
sns.boxplot(df.SepalLengthCm)
```

### The above is all univariate analysis

```python
df.shape
```

```python
colors = np.array(['red', 'blue', 'yellow'])
```

```python
plt.scatter(df.SepalLengthCm, np.zeros(150), color=colors[df.Species])
```

```python
df_temp = df[(df.Species == 1) | (df.Species == 0)]
```

```python
df_temp.shape
```

```python
plt.scatter(df_temp.SepalLengthCm, np.zeros(100), color=colors[df_temp.Species])
```
