```python
df[(df.mpg >20) & (df.hp > 110)]
```

```python
# transmission is automatic and milage > 20
df [ (df.am == 1) & (df.mpg > 20)] # always use bitwise & operator and not "and" operator
```

```python
# hp > 40 and mpg > 18 and am == 1
df[ (df.hp > 140) & (df.mpg > 18) & (df.am == 1) ]
```

# __SORTING THE DATAFRAME__

```python
df.sort_values('cyl', ascending=True, kind='mergesort', na_position='first')
```

```python
# you can see above that the index is jumbled as it moved with the record while ordering the data or sorting
# if one wants to ignore this you can use ignore_index kwarg
df.sort_values('cyl', ascending=True, kind='mergesort', na_position='first', ignore_index=True)
```

```python
# sorting using more than one column
df.sort_values(['model','mpg','hp'])
```

```python
# sort with more than one column where one column should be in ascending and other in descending
df.sort_values(['mpg','hp'], ascending=[True, False])
```
