# Pandas Group

```python
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

***Are automatic cars more economical than their manual counterparts ?***

# Grouping values

```python
cars_g1 = df.groupby(df.am)
cars_g1.mean()
```

```python
cars_g1.corrwith(df.mpg)
```

```python
cars_g1.median()
```

Assumption :
    More number of gears means less milage

```python
cars_g2 = df.groupby(df.gear)
cars_g2.mean()
```

Hence the assumption is false as more number of gears has a mean mpg > mean mpg for less number of gears

```python
cars_g2.median()
```
