## __iloc and loc__

```python
# iloc -> index location

#syntax df.iloc[row0:rowN, col0:colN]
df.iloc[0:2,:]
```

```python
# Syntax df.loc[row0:rowN:stepwise, 'colname1':'colnameN':stepwise]
df.loc[0:2,'model':'cyl':2]
```

```python
df.iloc[:5,1:7:2]
```

```python
df
```

```python
plt.scatter(df.mpg, df.disp)
```

```python
# using iloc select cadicall fleetwood, lotus europa, volvo142
# cols model, hp, wt
# iloc can show randomly chosen index and columns when provided in the form of a python list or an array
df.iloc[[14,27,31],[0,4,6]]
```

```python
# same goes for loc
df.loc[[14,27,31],['model','hp','wt']]
```

```python
plt.contour([df.hp, df.mpg])
```

```python
df.iloc[3:25, 1:11]
```

```python
# in loc last index is included both for rows and column even if there is a numerical index on either
df.loc[3:24,'mpg':'gear']
```
