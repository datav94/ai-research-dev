# Object Oriented Plots

Steps:

- Create the canvas
- add axes
- generate the plot
- specify the parameters

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
x = list(range(0,10))
y = [2,5,7,4,7,5,4,3,5,7]
```

```python
# create the canvas
fig = plt.figure()

# add axes
fig.add_axes([0,0,0.5,1]) # syntax fig.add_axes([x1,y1,x2,y2])

# generate the plot
plt.plot(x,y)
plt.show()
```

```python
# create the canvas
fig = plt.figure()

# add axes
fig.add_axes([0,0,1,1])

# generate the plot
plt.plot(x,y)
plt.show()
```

```python
# create the canvas
fig = plt.figure()

# add axes
ax = fig.add_axes([0,0,1,1])
ax.set_xlim(0,4)

# generate the plot
plt.plot(x,y)
plt.show()
```

```python
# create the canvas
fig = plt.figure()

# add axes
ax = fig.add_axes([0,0,1,1])
ax.set_xlim(0,4)
ax.set_ylim(4,8)

# generate the plot
plt.plot(x,y)
plt.show()
```

```python
%matplotlib inline # for older version of anaconda. Not needed right now.
fig = plt.figure()

# add axes
ax = fig.add_axes([0,0,1,1])
ax.set_xlim(1,5)
ax.set_ylim(3.4,8)

# generate the plot
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()

# add axes
ax = fig.add_axes([0,0,1,1])
ax.set_xlim(1,5)
ax.set_ylim(3,8)

ax.set_xticks([2,3,3.5]) # only shows those points on the graph x -axis
ax.set_yticks([4,5]) # only shows those points on the graph y-axis

# generate the plot
plt.plot(x,y)
plt.show()
```

# Formatting plots

```python
print(x,y)
y2 = [3,4,2,6,8,4,9,4,8,6]
```

```python
plt.plot(x,y, lw=3, marker='o',linestyle = ':', color='red', markersize=12)
plt.plot(x,y2, lw=3, marker='*',linestyle = '-.', color='green', markersize=12)
plt.title('customization of graphs')
plt.show()
```

```python
x = [0,1,2,3,4,5]
y = [4,5,1,6,8,9]
plt.bar(x,y, width = [0.3,1.5,0.8,2,2,1.5], color=['red','blue','pink','cyan','yellow','green']\
       ,edgecolor=['black', 'green','maroon'], clip_on=True)
plt.show()
```

```python
x = [0,1,2,3,4]
y = [1,2,3,4,5]
plt.bar(x,y, color=['red','blue'], width=[0.,0.7,0.8,0.9,1])
plt.show()
```
