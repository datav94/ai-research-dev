# save graphs

```python
# a histogram is a graph of frequency of occurance of each datapoint of a data
# example is as below
data = [1,1,2,2,5,5,5,5,6,7,7,7,7,4,4,4,3,3,3,3,3,3,3,3,3,3,3,8,8,9,9,9,10,10,10,10]

plt.hist(data, bins=9)
```

```python
sns.histplot(data, bins=9)
```

```python
sns.distplot(data, bins=5)
```

Bins are used to group data togather we have mentioned that bins=5 and our data is spread from 1 to 10
so what this bins does is that it divides the data into 5 groups so 1 and 2 is group 1, 3-4 is group 2 so on and so forth

```python
plt.hist(data)
plt.xlabel('something')
plt.ylabel('something else')
plt.title('another thing')
plt.show()
```

```python
# Draw multiple plots in a single figure
# make sure to type subplots and not subplot
rcParams['figure.figsize'] = [15,10]
fig, ax = plt.subplots(2,2) # we need 4 charts and hence 2 rows and 2 columns and hence 2,2

ax[0,0].plot(x,y)
ax[0,1].pie(data2,autopct='%0.2f', )
ax[1,0].bar(x,y)
ax[1,1].hist(data)

plt.show()
```
