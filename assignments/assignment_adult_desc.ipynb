```python
df_new = pd.read_csv('/home/bunny/CDS/pandas/adult.data', na_values=' ?')
df_new.columns = ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','annual_sal']

def normalize_to_digits(series):
    unique_data = list(series.unique())
    if np.nan in unique_data:
        unique_data.remove(np.nan)
    ser_dict = [{unique_data[i],i+1} for i in range(len(unique_data))]
    series.replace(unique_data, range(1,len(unique_data)+1), inplace=True)
    return ser_dict, series

digit_data = []
for series in df_new:
    if df_new[series].dtype == 'object':
        normalized = normalize_to_digits(df_new[series])
        digit_data.append(normalized[0])
```
